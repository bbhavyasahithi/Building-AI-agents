```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
loader=PyPDFLoader("monte carlo.pdf")
docs=loader.load()
print("Total Pages:",len(docs))
```

```python
sp=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
ds=sp.split_documents(docs)
print("Chunks:",len(ds))
```

```python
texts=[dc.page_content for dc in ds]
```

```python
vec=TfidfVectorizer()
emb=vec.fit_transform(texts)
print(emb.shape)
```

```python
vec_db=emb
print("Vector DB Created")
```

```python
query=input("Ask Question: ")
```

```python
query_vec=vec.transform([query])
```

```python
sim=cosine_similarity(query_vec,vec_db)
ind= sim[0].argsort()[-3:][::-1]
ind
```

```python
context=[]
for idx in ind:
    context.append(texts[idx])
    print("\n------------------")
    print(texts[idx][:500])
```

```python
prompt=f"""
Answer the question using the context below.
Context:
{' '.join(context)}
Question:
{query}
Answer:
"""
```

```python
answer="\n".join(context)
print(answer)
```
